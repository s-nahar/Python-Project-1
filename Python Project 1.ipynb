```python
import pandas as pd
import numpy as np
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('realtor-data.csv')
data.info()
data.head
```

```python
X = data[['bed','bath','acre_lot', 'house_size']]
Y = data['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```
